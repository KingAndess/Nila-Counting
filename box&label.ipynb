```python
from ultralytics import YOLO
import cv2

# Load model
model = YOLO('/home/andes/skripsi/kodingan2/runs/train/exp12/weights/best.pt')  # Path ke model best.pt yang sudah Anda latih

# Baca video
video_path = '/home/andes/skripsi/kodingan2/Test-Vid/17-1.mp4'  # Ganti dengan path video yang ingin diuji
cap = cv2.VideoCapture(video_path)

# Mendapatkan ukuran frame untuk output video
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Output video
out = cv2.VideoWriter('Label.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Melakukan inferensi pada setiap frame
    results = model.predict(frame, save=False)  # Jangan simpan hasil otomatis, cukup prediksi

    # Menggambar kotak prediksi pada frame
    annotated_frame = results[0].plot()  # Menggambar kotak deteksi

    # Menyimpan frame yang sudah diannotasi ke output video
    out.write(annotated_frame)

    # Tampilkan hasil secara langsung (opsional)
    cv2.imshow('YOLOv10 Inference', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Melepaskan resource
cap.release()
out.release()
cv2.destroyAllWindows()
```
