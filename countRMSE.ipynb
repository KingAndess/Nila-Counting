```python
import cv2
import numpy as np
from ultralytics import YOLO

# Load YOLO model
model = YOLO('/home/andes/skripsi/Nila-Counting/runs/train/E100B16/weights/best.pt')  # Ganti dengan path ke best.pt

# Ground truth (jumlah bibit ikan yang sebenarnya dalam video)
ground_truth = 200  # Ganti dengan jumlah bibit ikan sebenarnya dalam video tersebut

# Load video
video_path = './Test-Vid/200.mp4'  # Ganti dengan path video yang sesuai
cap = cv2.VideoCapture(video_path)

# Variabel untuk menghitung jumlah deteksi dan jumlah frame
total_count = 0
frame_count = 0

detected_per_frame = []

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Melakukan inferensi pada setiap frame
    results = model.predict(frame, save=False)  # Prediksi tanpa menyimpan otomatis
    
    # Hitung jumlah deteksi (jumlah bounding box yang ditemukan)
    count = 0
    for result in results:
        count += len(result.boxes)  # Hitung jumlah box deteksi

    detected_per_frame.append(count)

    total_count += count
    frame_count += 1

# Melepaskan resource
cap.release()

# Hitung rata-rata jumlah bibit per frame
average_detected = total_count / frame_count if frame_count > 0 else 0

# Menghitung RMSE
# Jika ground_truth adalah jumlah bibit ikan sebenarnya, maka kita hitung RMSE sebagai berikut
detected_per_frame = np.array(detected_per_frame)
rmse = np.sqrt(np.mean((detected_per_frame - ground_truth)**2))

print(f"Ground Truth: {ground_truth}")
print(f"Rata-rata Deteksi: {average_detected}")
print(f"RMSE: {rmse:.2f}")
```
